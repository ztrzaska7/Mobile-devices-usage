```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
data=pd.read_csv('user_behavior_dataset.csv')
print(data.head())
```

```python
print(data.isnull().sum())
```

## Age distribution of phone users

```python
import seaborn as sns
plt.figure(figsize=(10,5))
sns.histplot(data['Age'],bins=20,kde=True,color='red')
plt.title('Distribution of Users age')
plt.xlabel("Age")
plt.ylabel('Frequency')
plt.show()
```

```python
#Gender distribution
plt.figure(figsize=(10,6))
gender_counts=data['Gender'].value_counts()
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', startangle=140, colors=['#ff9999','#66b3ff'])
plt.title('Gender Distribution')
plt.axis('equal')
plt.show()
```

```python
# distribution of battery drain
```

```python
plt.figure(figsize=(10,5))
sns.histplot(data['Battery Drain (mAh/day)'],kde=True)
plt.title('Distribution of Battery Drain (mAH/day)')
plt.xlabel('Battery Drain')
plt.ylabel('Frequency')
plt.show()
```

```python
#Now Let's take a look at distribution of device models
one=round((data['Device Model'].value_counts()/data.shape[0])*100,2).reset_index()
labels=one['Device Model'].to_list()
plt.pie(one['count'],labels=labels,autopct='%0.2f%%')
plt.title('Percentage Usage distribution of Device Model',font='16')
plt.show()
```
